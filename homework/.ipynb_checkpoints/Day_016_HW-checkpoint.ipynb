```python
# 載入需要的套件
import os
import numpy as np 
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

### 仿造之前做過的處理

```python
# 設定 data_path
dir_data = '~/project/100Day-ML-Marathon/data/'
f_app_train = os.path.join(dir_data, 'application_train.csv')
f_app_test = os.path.join(dir_data, 'application_test.csv')

# 讀取檔案
app_train = pd.read_csv(f_app_train)
app_test = pd.read_csv(f_app_test)

app_train
```

# - 原始資料是有序離散值的話 => Label Encoding
# - 原始資料是無序離散值的話 => One Hot Encoding (Dummies)

```python
# 種類 2 種以下的類別型欄位轉標籤編碼 (Label Encoding)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le_count = 0
le_col = []

# 檢查 dtype
for col in app_train:
    if app_train[col].dtype == 'object':
        # 如果只有兩種值的類別型欄位
        if len(list(app_train[col].unique())) <= 2:
            # 就做 Label Encoder
            le.fit(app_train[col])
            app_train[col] = le.transform(app_train[col])
            app_test[col] = le.transform(app_test[col])
            
            # 紀錄有多少個 columns 被標籤編碼過
            le_count += 1
            print(col)
            le_col.append(col)
            
            
```

```python
# 標籤編碼 (2種類別) 欄位轉 One Hot Encoding            
app_train = pd.get_dummies(app_train)
app_test = pd.get_dummies(app_test)
```

# 2020 10 23 心得：
- 1. 先用 Label Encoding 把兩種以內的文字轉數字
- 2. 多值 大於2種以上，就用 pd.get_dummies() 做 One Hot Encoding，把每個值配上column name 做成獨立欄位，並填上值，避免有排序的情形發生
